```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load data

```python
BCData = pd.read_csv("wisc_bc_data.csv")
```

```python
BCData.head(10)
```

```python
BCData.dtypes
```

```python

BCData['diagnosis'] = BCData.diagnosis.astype('category')
```

```python
BCData.head()
```

```python
BCData.describe().transpose()
```

```python
BCData.groupby(["diagnosis"]).count()
```

```python
BCData = BCData.drop(labels = "id", axis = 1)
```

```python
BCData.shape
```

```python
X = BCData.drop(labels= "diagnosis" , axis = 1)
y = BCData["diagnosis"]
X.head()
```

```python
#convert the features into z scores
XScaled  = X.apply(zscore)
XScaled.describe()
```

## Split data

```python
X_train, X_test, y_train, y_test = train_test_split(XScaled, y, test_size=0.30, random_state=1)
```

## KNN model

```python
model = KNeighborsClassifier(n_neighbors= 9 , weights = 'distance' ,metric='euclidean')
```

```python
model.fit(X_train, y_train)              #training model
```

## Evaluating model

```python
model.score(X_test, y_test)
```

## Confusion matrix

```python
from sklearn import metrics
prediction = model.predict(X_test)
print("Confusion Matrix")
cm=metrics.confusion_matrix(y_test, prediction, labels=["M", "B"])

df_cm = pd.DataFrame(cm, index = [i for i in ["M","B"]],
                  columns = [i for i in ["Predict M","Predict B"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True,fmt='g')
```
